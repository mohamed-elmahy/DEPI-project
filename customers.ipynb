```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/customers.csv")
data.head(10)
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data['last_name'] = data['last_name'].str.replace(r'[-,@1-9]+$', '', regex=True)
data['last_name'] = data['last_name'].str.title()
display(data.head(10))
```

```python
data['first_name'] = data['first_name'].str.replace(r'[-,@1-9]+$', '', regex=True)
data['first_name'] = data['first_name'].str.title()
display(data.head(10))
```

```python
data['email'] = data['email'].fillna("No mail")
data['phone'] = data['phone'].fillna("No phone")
data['signup_date'] = data['signup_date'].fillna("No signup_date")
data.head(20)
```
